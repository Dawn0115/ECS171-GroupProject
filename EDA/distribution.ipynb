```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df_train = pd.read_csv("train_cleaned.csv")


numerical_cols = df_train.select_dtypes(include='number').columns.tolist()
for col_to_drop in ['id']:
    if col_to_drop in numerical_cols:
        numerical_cols.remove(col_to_drop)


sns.set(style="whitegrid")


for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.histplot(df_train[col], bins=30, kde=True, color="#f4a261", edgecolor='black', alpha=0.6)

    plt.title(f"Histogram for {col}", fontsize=14, weight='bold')
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f"train_dist_{col}.png")  # 
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df_train = pd.read_csv("train_cleaned.csv")


sex_counts = df_train["Sex"].value_counts()
sex_percent = df_train["Sex"].value_counts(normalize=True) * 100


colors = ["#7cb342", "#0288d1", "#cddc39"] 


plt.figure(figsize=(6, 6))
wedges, texts, autotexts = plt.pie(
    sex_counts,
    labels=sex_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    textprops={'color': "black"},
    wedgeprops=dict(width=0.4)  # 
)


plt.gca().set_aspect('equal')
plt.title("Pie Chart for Sex", fontsize=14, weight='bold')
plt.tight_layout()
plt.savefig("donut_chart_sex.png")
plt.show()
```
