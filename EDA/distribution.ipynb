```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math


df_train = pd.read_csv("train_cleaned.csv")


numerical_cols = df_train.select_dtypes(include='number').columns.tolist()
numerical_cols = [col for col in numerical_cols if col != 'id']


sns.set(style="whitegrid")


n_cols = 3
n_rows = math.ceil(len(numerical_cols) / n_cols)
fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 5, n_rows * 4))


axes = axes.flatten()

for i, col in enumerate(numerical_cols):
    sns.histplot(df_train[col], bins=30, kde=True, color="#f4a261", edgecolor='black', alpha=0.6, ax=axes[i])
    axes[i].set_title(f"{col}", fontsize=12, weight='bold')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Count")


for j in range(len(numerical_cols), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.savefig("train_all_histograms.png") 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df_train = pd.read_csv("train_cleaned.csv")


sex_counts = df_train["Sex"].value_counts()
sex_percent = df_train["Sex"].value_counts(normalize=True) * 100


colors = ["#7cb342", "#0288d1", "#cddc39"] 


plt.figure(figsize=(6, 6))
wedges, texts, autotexts = plt.pie(
    sex_counts,
    labels=sex_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    textprops={'color': "black"},
    wedgeprops=dict(width=0.4)  # 控制内圈大小
)


plt.gca().set_aspect('equal')
plt.title("Pie Chart for Sex", fontsize=14, weight='bold')
plt.tight_layout()
plt.savefig("donut_chart_sex.png")
plt.show()
```
